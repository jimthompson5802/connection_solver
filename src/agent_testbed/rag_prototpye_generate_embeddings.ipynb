```python
import pickle
import json
import os
```

```python
with open("/openai/api_key.json") as f:
    config = json.load(f)


os.environ["OPENAI_API_KEY"] = config["key"]
```

```python

with open('vocabulary.pkl', 'rb') as file:
    vocabulary = pickle.load(file)
```

```python
# print the vocabulary
for word, definitions in vocabulary.items():
    print(f"\n\n{word}:")
    for definition in definitions:
        print(definition)
```

```python
from openai import OpenAI

client = OpenAI()

# Function to get embedding for a sentence
def get_embedding(sentence):
    response = client.embeddings.create(
        input=[sentence],
        model="text-embedding-3-small"
    )
    return response.data[0].embedding

# Create a dictionary to store embeddings for each word and its definitions
embeddings = {}

for word, definitions in vocabulary.items():
    embeddings[word] = [get_embedding(definition) for definition in definitions]

# Print the embeddings for the word 'throw'
print(embeddings['throw'])
```

```python
with open('embeddings.pkl', 'wb') as file:
    pickle.dump(embeddings, file)
```
