```python
import pickle
import json

import pandas as pd
```

```python
pickled_fp = 'snapshot_testbed.pkl'

# read in the pickled data
with open(pickled_fp, 'rb') as f:
    data = pickle.load(f)

data[:5]
```

```python
keys_to_keep = ['v','ts', 'id', 'channel_values']
data_to_extract = [{k:v for k,v in d.checkpoint.items() if k in keys_to_keep} for d in data]
```

```python
df = pd.json_normalize(data_to_extract).sort_values('ts')
```

```python
df.dtypes
```

```python
# keep only columns names that do not have the string "branch" or "start"
columns_to_keep = [col for col in df.columns if 'branch' not in col and 'start' not in col]
columns_to_keep
```
