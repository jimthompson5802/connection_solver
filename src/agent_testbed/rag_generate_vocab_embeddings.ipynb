```python
import json
import os
import pickle
import sys

import pandas as pd
import numpy as np

# Import the 'rag_tools' module from the 'agent' directory
current_dir = os.getcwd()
project_dir = os.path.dirname(current_dir)

# Add the 'agent' directory to the sys.path
other_dir = os.path.join(project_dir, 'agent')
sys.path.append(other_dir)

# confirm the 'agent' directory is in the sys.path
for p in sys.path:
    print(p)

from embedvec_tools import generate_vocabulary, generate_embeddings
```

```python
with open("/openai/api_key.json") as f:
    config = json.load(f)


os.environ["OPENAI_API_KEY"] = config["key"]
```

```python
pickle_fp = "word_list6.pkl"
words = ["grain", "drain", "total", "cube", "syrup", "shred", "signature", "sap", "jam", "tax", "powder", "rock", "empty", "groove", "tip", "exhaust"]

vocabulary = generate_vocabulary(words)


# for word, definitions in vocabulary.items():
#     print(f"\n\n{word}:")
#     for definition in definitions:
#         print(definition)

# Convert dictionary to DataFrame
rows = []
for word, definitions in vocabulary.items():
    for definition in definitions:
        rows.append({'word': word, 'definition': definition})

df = pd.DataFrame(rows)

# Generate embeddings
embeddings = generate_embeddings(df["definition"].tolist())
df["embedding"] = [np.array(v) for v in np.array(embeddings).tolist()]

# save the dataframe as a pickle file
df.to_pickle(pickle_fp)

df
```
